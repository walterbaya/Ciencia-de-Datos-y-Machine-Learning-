# Crear un subconjunto de datos

Extraigo la columna Account Length
Este objeto es de tipo series, el cual consiste en los valores de la columna

```python
account_length = data["Account Length"]
type(account_length)
```

Seleccionamos mas de una columna
Python no puede autocompletar esto por nosotros

```python
subset = data[["Account Length", "Phone", "Eve Charge", "Day Calls"]]
```

Hemos obtenido un subconjunto de los datos
Vemos que como son multiples columnas automaticamente se convierte en un Dataframe

```python
subset.head()
type(subset)
```

Podemos escribir que columnas deseamos previamente
Tener muchas columnas que no son necesarias ocupa espacio, por lo tanto
Es preferible tener solo lo necesario siempre y no mas.

```python
desired_columns = ["Account Length", "Phone", "Eve Charge", "Night Calls"]
subset = data[desired_columns]
subset.head()
```

Queremos quedarnos con el complementario

```python
desired_columns = ["Account Length", "VMail Message", "Day Calls"]
all_columns_list = data.columns.values.tolist()
```

En sublist estaran todas las que no quiero

```python
sublist = [x for x in all_columns_list if x not in desired_columns]
```

```python
subset = data[sublist]
subset.head()
```

Otra manera mas linda 

```python
a = set(desired_columns)
b = set(all_columns_list)
sublist = b-a
sublist = list(sublist)
```

```python
##Usuarios con total mins superior a 500
#Filtrado por booleanos
data1 = data[data["Day Mins"]>300]
data1
```

```python
##Usuarios de Nueva York()State = "NY"
data2 = data[data["State"]=="NY"]
data2
```

```python
##AND -> & 
data3 = data[(data["Day Mins"]>300) & (data["State"] == "NY")]
data3
##OR -> |
data4 = data[(data["Day Mins"]>300) | (data["State"] == "NY")]
data4
```
